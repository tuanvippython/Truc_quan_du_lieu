```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/ADMIN/Downloads/california_cities.csv")
df
```

```python
# chọn kiểu biểu đồ
plt.style.available

plt.style.use("seaborn-whitegrid")
```

# Biểu đồ Scatter

```python
latd,longd = df[["latd"]] , df[["longd"]]


# Vẽ biểu đồ 
fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(longd,latd,color ="blue")

# chú thích các giá trị cho biểu đồ
ax.set(title = "Biểu đồ thể hiện tọa độ của các thành phố trong tiểu bang Cali",
      xlabel = "Kinh độ (°)",
      ylabel = "Vĩ độ (°)");

ax.title.set_fontsize(20) 
ax.xaxis.label.set_fontsize(18)  
ax.yaxis.label.set_fontsize(18)

ax.tick_params(axis='x', labelsize=16) 
ax.tick_params(axis='y', labelsize=16)  
```

# Biểu đồ Histogram

```python


# Lấy ra 100 dữ liệu dân số đầu tiên
populations = df["population_total"].head(100)

fig,ax = plt.subplots(figsize=(12, 7))
ax.hist(populations, color='skyblue',bins=40)

# chú thích các giá trị cho biểu đồ
ax.set(title ="Biểu đồ thể hiện dân số của 100 thành phố đầu tiên trong dữ liệu",
      xlabel ="Dân số (người)",
      ylabel = "Tần suất");

ax.title.set_fontsize(20) 
ax.xaxis.label.set_fontsize(18)  
ax.yaxis.label.set_fontsize(18)

ax.tick_params(axis='x', labelsize=16) 
ax.tick_params(axis='y', labelsize=16)  
```

# Biểu đồ đường

```python
df.index

# dữ liệu độ cao theo mét và độ cao theo feet
elevation_m = df['elevation_m']
elevation_ft = df['elevation_ft']

# vẽ biểu đồ
fig,ax= plt.subplots(figsize=(12,8))

ax.plot(df.index, elevation_m, label='Elevation (m)')
ax.plot(df.index, elevation_ft, label='Elevation (ft)')

# chú thích các giá trị cho biểu đồ
ax.set(title ="Biểu đồ đường độ cao của thành phố",
      xlabel = "Các thành phố",
      ylabel = "Độ cao")

ax.title.set_fontsize(20) 
ax.xaxis.label.set_fontsize(18)  
ax.yaxis.label.set_fontsize(18)

ax.tick_params(axis='x', labelsize=16) 
ax.tick_params(axis='y', labelsize=16)  

ax.legend(fontsize=18);
```

# Biểu đồ mật độ

```python
df["population_density"] = df["population_total"] / df["area_land_km2"]
df["population_density"].head(100)
```

```python

# Tính mật độ dân số (người/km^2)
df["population_density"] = df["population_total"] / df["area_land_km2"]

fig,ax = plt.subplots(figsize = (12,8))
sns.kdeplot(df["population_density"].head(100), fill=True, color='skyblue')

ax.set(title ="Biểu đồ của Mật độ dân số cho 100 thành phố đầu tiên trong dữ liệu",
      xlabel ="Mật độ dân số (người/km^2)",
      ylabel ="Tỉ trọng" )


ax.xaxis.label.set_fontsize(18)  
ax.yaxis.label.set_fontsize(18)

ax.tick_params(axis='x', labelsize=16) 
ax.tick_params(axis='y', labelsize=16)  
```

# Biểu đồ Doughnut

```python

# Lấy dữ liệu từ dataframe
elevation_water_percent = df["area_water_percent"].mean()
elevation_land_percent = 100 - elevation_water_percent

# Tạo dữ liệu cho biểu đồ
sizes = [elevation_land_percent, elevation_water_percent]
labels = ['Đất liền', 'Nước']

# tạo biểu đồ doughnut
fig, ax = plt.subplots(figsize = (12,8))
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['#FF9999', '#66B2FF'],textprops={'fontsize': 16})
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Chú thích cho biểu đồ
ax.set(title = "Biểu đồ phần trăm đất liền và nước trong các thành phố ")
ax.title.set_fontsize(20) 


```
