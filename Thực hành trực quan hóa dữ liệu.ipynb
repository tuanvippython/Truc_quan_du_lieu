```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Biểu đồ Scatter

```python
latd,longd = df[["latd"]] , df[["longd"]]


# Vẽ biểu đồ 
fig,ax = plt.subplots(figsize=(12,8))
ax.scatter(longd,latd,color ="blue")

# chú thích các giá trị cho biểu đồ
ax.set(title = "Biểu đồ thể hiện tọa độ của các thành phố trong tiểu bang Cali",
      xlabel = "Kinh độ (°)",
      ylabel = "Vĩ độ (°)");

ax.title.set_fontsize(20) 
ax.xaxis.label.set_fontsize(18)  
ax.yaxis.label.set_fontsize(18)

ax.tick_params(axis='x', labelsize=16) 
ax.tick_params(axis='y', labelsize=16)  
```

# Biểu đồ mật độ

```python
df["population_density"] = df["population_total"] / df["area_land_km2"]
df["population_density"].head(100)
```

```python

# Tính mật độ dân số (người/km^2)
df["population_density"] = df["population_total"] / df["area_land_km2"]

fig,ax = plt.subplots(figsize = (12,8))
sns.kdeplot(df["population_density"].head(100), fill=True, color='skyblue')

ax.set(title ="Biểu đồ của Mật độ dân số cho 100 thành phố đầu tiên trong dữ liệu",
      xlabel ="Mật độ dân số (người/km^2)",
      ylabel ="Tỉ trọng" )
ax.title.set_fontsize(20) 

ax.xaxis.label.set_fontsize(18)  
ax.yaxis.label.set_fontsize(18)

ax.tick_params(axis='x', labelsize=16) 
ax.tick_params(axis='y', labelsize=16)  
```

# Biểu đồ Doughnut

```python

# Lấy dữ liệu từ dataframe
elevation_water_percent = df["area_water_percent"].mean()
elevation_land_percent = 100 - elevation_water_percent

# Tạo dữ liệu cho biểu đồ
sizes = [elevation_land_percent, elevation_water_percent]
labels = ['Đất liền', 'Nước']

# tạo biểu đồ doughnut
fig, ax = plt.subplots(figsize = (12,8))
ax.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['#FF9999', '#66B2FF'],textprops={'fontsize': 16})
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Chú thích cho biểu đồ
ax.set(title = "Biểu đồ diện tích đất liền và nước trong các thành phố ")
ax.title.set_fontsize(20) 


```

# Biểu đồ cột

```python

# Lấy 20 thành phố đầu tiên
cities = df.iloc[:20, 1]
population = df.iloc[:20, 6]

# Tạo biểu đồ cột
fig, ax = plt.subplots(figsize=(15, 8))

bars = ax.bar(cities, population, label="Dân số (người)", color='skyblue')

# Thêm chỉ số vào đỉnh từng cột
for bar, index in zip(bars, population):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05,
             str(index), ha='center', va='bottom', fontsize=14)

ax.set(title='Biểu đồ thể hiện dân số của 20 thành phố đầu tiên trong dữ liệu',
       ylabel='Dân số (người)')

plt.xticks(rotation=45, ha='right', fontsize=14)
ax.title.set_fontsize(20)
ax.xaxis.label.set_fontsize(18)
ax.yaxis.label.set_fontsize(18)

ax.tick_params(axis='x', labelsize=16)
ax.tick_params(axis='y', labelsize=16)

ax.legend(fontsize=18)

plt.show()
```

```python
# Lấy 10 thành phố đầu tiên
cities = df.iloc[:10, 1]
area_total_sq_mi = df.iloc[:10, 7].round(3)
area_land_sq_mi = df.iloc[:10, 8].round(3)

fig, ax = plt.subplots(figsize=(18, 16))

bar_width = 0.4
bar1 = ax.bar(cities.index, area_total_sq_mi, width=bar_width, label='Diện tích tổng (Km^2)', color='skyblue')
bar2 = ax.bar(cities.index + bar_width, area_land_sq_mi, width=bar_width, label='Diện tích đất (Km^2)', color='orange')

# Thêm chỉ số lên đỉnh của cột Diện tích tổng
for bar, index in zip(bar1, area_total_sq_mi):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1,
            str(index), ha='center', va='bottom', rotation=45, fontsize=14)

# Thêm chỉ số lên đỉnh của cột Diện tích đất
for bar, index in zip(bar2, area_land_sq_mi):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.1,
            str(index), ha='center', va='bottom', rotation=45, fontsize=14)

ax.set(title ='Biểu đồ thể hiện diện tích tổng và Diện tích đất của 10 thành phố đầu tiên trong dữ liệu',
      xlabel='Thành phố',
      ylabel ='Diện tích (km^2)')   



ax.title.set_fontsize(22) 
ax.xaxis.label.set_fontsize(20)  
ax.yaxis.label.set_fontsize(20)

ax.tick_params(axis='x', labelsize=18) 
ax.tick_params(axis='y', labelsize=18)  
    

ax.set_xticks(cities.index + bar_width / 2)
ax.set_xticklabels(cities, rotation=45, ha='right')
ax.legend(loc='upper center',fontsize = 22);


```

# Biểu đồ tròn

```python
# Lấy 10 thành phố ngẫu nhiên từ DataFrame
np.random.seed(42)
random_cities = df.sample(n=10)

# Lấy cột dân số cho giá trị biểu đồ
values = random_cities['population_total']
labels = random_cities['city']

# Thiết lập kích thước biểu đồ
plt.figure(figsize=(10, 10))

# Vẽ biểu đồ tròn
plt.pie(values, labels=labels, autopct="%1.1f%%", startangle=100)

# Thêm tiêu đề
plt.title("Biểu đồ thể hiện dân số của 10 thành phố ngẫu nhiên", fontsize=16)

# Hiển thị biểu đồ
plt.show()
```

# Biểu đồ vùng

```python
# Lấy 10 thành phố đầu tiên
df_subset = df.head(20)
cities = df_subset['city']
area_land_sq_mi = df_subset['area_land_sq_mi']
area_water_sq_mi = df_subset['area_water_sq_mi']

# vẽ biểu đồ
fig,ax = plt.subplots(figsize=(12, 8))

ax.stackplot(cities, area_land_sq_mi, area_water_sq_mi, labels=['Diện tích đất (km^2)', 'Diện tích nước (km^2)'])
plt.xticks(rotation=90)
ax.set(title='Diện tích đất và nước của 20 các thành phố đầu tiên trong dữ liệu',
      xlabel='Các thành phố',
      ylabel='Diện tích đất và nước (km^2)')

ax.title.set_fontsize(20) 
ax.xaxis.label.set_fontsize(18)  
ax.yaxis.label.set_fontsize(18)

ax.tick_params(axis='x', labelsize=16) 
ax.tick_params(axis='y', labelsize=16)  


ax.legend(loc='upper left',fontsize=20)
plt.tight_layout()
```

# Biểu đồ nhiệt

```python

# Lấy thông tin về tên thành phố và lượng nước cho 10 thành phố đầu tiên
cities = df['city'].head(10)
water_area_percent = df['area_water_percent'].head(10)

heatmap_data = pd.DataFrame({'Thành phố': cities, 'Tỷ lệ Diện tích Mặt nước (%)': water_area_percent})

plt.figure(figsize=(12, 8))
ax = plt.gca()

# Vẽ biểu đồ nhiệt với mối quan hệ giữa lượng nước và thành phố
heatmap = sns.heatmap(heatmap_data.set_index('Thành phố'), cmap='coolwarm', annot=True, fmt='.2f', ax=ax, cbar_kws={'label': 'Đơn vị (%)'},
                      annot_kws={'size': 16})
# Thiết lập tiêu đề và hiển thị biểu đồ
plt.title('Biểu đồ nhiệt về Tỷ lệ Diện tích Mặt nước theo Thành phố', fontsize=18)

# Chỉnh kích thước font chữ của nhãn trục x và trục y
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

# Chỉnh kích thước font chữ của nhãn trục x và trục y
cbar = heatmap.collections[0].colorbar
cbar.ax.tick_params(labelsize=16)

plt.show()
```

# Biểu đồ bong bóng

```python
np.random.seed(2)
random_sample = df.sample(10)

# Trích xuất thông tin từ dữ liệu ngẫu nhiên
cities = random_sample["city"]
populations = random_sample["population_total"]
areas = random_sample["area_land_km2"]
elevations = random_sample["elevation_m"]

#Chỉnh kích thước bong bóng
bubble_size = elevations *15
# Vẽ biểu đồ bong bóng
fig,ax = plt.subplots(figsize=(14, 10))
ax.scatter(areas, populations, s=bubble_size, alpha=0.5, c='blue')

# Đặt tên cho trục và tiêu đề
ax.set(title="Biểu đồ thể hiện mối quan hệ giữa diện tích Đất, dân số và độ cao của Thành Phố",
      xlabel ='Diện tích đất (km²)',
      ylabel ='Dân số (người)')
# Hiển thị tên của từng thành phố
for i, city in enumerate(cities):
    plt.annotate(city, (areas.iloc[i], populations.iloc[i]), textcoords="offset points", xytext=(0, 5), ha='center',fontsize=16)

ax.title.set_fontsize(20) 
ax.xaxis.label.set_fontsize(18)  
ax.yaxis.label.set_fontsize(18)

ax.tick_params(axis='x', labelsize=16) 
ax.tick_params(axis='y', labelsize=16)  



```
